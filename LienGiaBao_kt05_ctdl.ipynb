```python
from collections import deque

def max_sliding_window(nums, k):
    dq = deque()  # Deque lưu chỉ số phần tử
    result = []   # Danh sách kết quả

    for i in range(len(nums)):
        # 1. Loại bỏ phần tử đã ra khỏi cửa sổ
        if dq and dq[0] < i - k + 1:
            dq.popleft()

        # 2. Loại bỏ phần tử nhỏ hơn nums[i] trong deque
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()

        # 3. Thêm phần tử hiện tại vào deque
        dq.append(i)

        # 4. Khi cửa sổ đạt kích thước k, lấy max từ phần tử đầu deque
        if i >= k - 1:
            result.append(nums[dq[0]])

    return result

# Test case
num_list = [3, 4, 5, 1, -44, 5, 10, 12, 33, 1]
k = 3
print(max_sliding_window(num_list, k))  # Output: [5, 5, 5, 5, 10, 12, 33, 33]
```

```python
def count_chars(word):
    char_count = {}  # Khởi tạo dictionary rỗng
    for char in word:  # Duyệt qua từng ký tự trong từ
        if char in char_count:
            char_count[char] += 1  # Tăng số lần xuất hiện nếu đã có
        else:
            char_count[char] = 1  # Thêm vào dictionary nếu chưa có
    return char_count

# Nhập từ bất kỳ từ người dùng
word = input("Nhập một từ: ")
result = count_chars(word)
print(result)
```

```python
import gdown
import collections

def download_and_count_words(file_id):
    # Link tải trực tiếp từ Google Drive
    gdrive_url = f"https://drive.google.com/uc?id={file_id}"
    output = "downloaded_file.txt"

    # Tải file về
    gdown.download(gdrive_url, output, quiet=False)

    # Đọc nội dung file
    with open(output, "r", encoding="utf-8") as f:
        text = f.read()

    # Chuyển chữ về lowercase và tách từ
    words = text.lower().split()

    # Đếm số lần xuất hiện của mỗi từ
    word_count = collections.Counter(words)

    return dict(word_count)

file_id = "16osHYfaezFcYUmW9z_ovupswsu7vO_Ny"
result = download_and_count_words(file_id)
print(result)


```

```python
def levenshtein_distance(source, target):
    m, n = len(source), len(target)

    # Khởi tạo ma trận (m+1) x (n+1)
    D = [[0] * (n + 1) for _ in range(m + 1)]

    # Điền hàng và cột đầu tiên
    for i in range(m + 1):
        D[i][0] = i  # Chi phí xóa
    for j in range(n + 1):
        D[0][j] = j  # Chi phí chèn

    # Tính toán các giá trị trong ma trận
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            delete_cost = D[i - 1][j] + 1  # Xóa ký tự
            insert_cost = D[i][j - 1] + 1  # Chèn ký tự
            sub_cost = D[i - 1][j - 1] + (0 if source[i - 1] == target[j - 1] else 1)  # Thay thế ký tự

            D[i][j] = min(delete_cost, insert_cost, sub_cost)

    return D[m][n]

# Nhập chuỗi từ bàn phím
source = input("Nhập chuỗi nguồn: ")
target = input("Nhập chuỗi đích: ")

# Tính khoảng cách Levenshtein
distance = levenshtein_distance(source, target)
print(f"Khoảng cách Levenshtein giữa '{source}' và '{target}' là {distance}")
```
